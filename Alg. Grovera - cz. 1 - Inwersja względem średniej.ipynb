## Algorytm wyszukiwania kwantowego (Algorytm Grovera)

- Dostp do listy uzyskujemy poprzez wyroczni: mo偶emy zadawa zapytania dotyczce zawartoci listy.

- Ka偶dy stan bazowy, tj. $|0 \cdots 0>, \ldots, |1 \cdots 1>$, odpowiada indeksowi listy.
- Jeli poszukiwany element zostanie wykryty, wyrocznia odwraca znak odpowiedniej amplitudy.


<i style="color:blue;">Zauwa偶, 偶e w powy偶szych grach symulujemy wyroczni, znajc wczeniej zaznaczone elementy. W rzeczywistej implementacji wyrocznia powinna mie mechanizm kwantowy wykrywajcy zaznaczone elementy, bdc w superpozycji wszystkich indeks贸w, a nastpnie znak odpowiedniej amplitudy zostanie automatycznie odwr贸cony. Na pocztku do ka偶dego kubitu stosowany jest operator Hadamarda. Zatem amplituda ka偶dego stanu bazowego jest ustawiona na $ \frac{1}{\sqrt{N}} $. Nastpnie w kolejnych iteracjach gry zmieniaj si amplitudy oznaczonych i nieoznaczonych element贸w.

<h4> Liczba iteracji </h4>

Gdy liczba oznaczonych element贸w jest mniejsza ni偶 element贸w nieoznaczonych, amplitudy oznaczonych element贸w zaczynaj rosn. 

Nastpnie osigamy pierwszy szczyt, w kt贸rym prawdopodobiestwo zaobserwowania zaznaczonego elementu przyjmuje maksymaln warto.

Po przekroczeniu tego punktu amplituda oznaczonych element贸w zaczyna spada.
